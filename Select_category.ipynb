# Getting data that matches specific categories

This is the notebook to query and generate dataframes based on different categories. 

Example:
We see that 'Drama' with id 18 is in the 'genres' column of the dataframe, and we would only want rows with the specific genre. The function needed will be:


```
retrieve_category(dataframe=df, column_name='genres', id='18', id_key_name='id', result_file_name='drama_related_movies')
```
It thens returns the desired dataframe and save the result to a pickled file.


# Loading the cleaned version of the data

````python
data_dir = 'data/pre-processed/'
````

````python
import pandas as pd

movie_details_neat = pd.read_pickle(data_dir + 'movie_details_neat.pkl')

movie_details_neat.head()
````

# Select Function

````python
def retrieve_category(df, column_name, ids, id_key_name, result_file_name=''):
    result = df[df.apply(lambda row: False if id_key_name not in row[column_name].keys() else (ids in (row[column_name][id_key_name])), axis=1)]
    if not result_file_name:
        result_file_name = column_name + '_' + str(ids)
    result.to_pickle(data_dir + result_file_name + '.pkl')
    return result
````

# Example 


````python
drama_rows = retrieve_category(movie_details_neat, 'genres', 18, 'id', '')

drama_rows.head()
````
